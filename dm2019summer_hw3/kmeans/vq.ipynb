Run your k-means algorithm on `kmeans_data.mat`. Note that the lines indicate the iterations of centroids.

```python
from kmeans import kmeans
from kmeans_plot import kmeans_plot
x = sio.loadmat('kmeans_data.mat')['X']
N, P = x.shape
k = 2  # modify if you want
max_iter = 1000
min_dist = float('inf')
max_dist = -float('inf')
max_idx = np.zeros(N)
min_idx = np.zeros(N)
max_ctrs = np.zeros((k, P))
max_iter_ctrs = np.zeros((max_iter, k, P))
min_ctrs = np.zeros((k, P))
min_iter_ctrs = np.zeros((max_iter, k, P))
for i in range(1000):
    idx, ctrs, iter_ctrs, dist = kmeans(x, k)
    if(dist>max_dist):
        max_dist = dist
        max_idx = idx
        max_ctrs = ctrs
        max_iter_ctrs = iter_ctrs
    if(dist<min_dist):
        min_dist = dist
        min_idx = idx
        min_ctrs = ctrs
        min_iter_ctrs = iter_ctrs
        
kmeans_plot(x, max_idx, max_ctrs, max_iter_ctrs)
kmeans_plot(x, min_idx, min_ctrs, min_iter_ctrs)
```
